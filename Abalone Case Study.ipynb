```python
#Creating train-test split

from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score

from sklearn import metrics
```

```python
###Feature Selection
##Feature Selection is a techinque of finding out the features that contribute the most to our model 
#i.e. the best predictors.


x_train, x_test, y_train, y_test = train_test_split(x,y,random_state = None)
print('Training set shape: ', x_train.shape, y_train.shape)
print('Testing set shape: ', x_test.shape, y_test.shape)
```

```python
# Spot Checking and Comparing Algorithms 
n_neighbors=5
models=[]

##importing libraries
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor

#appending models to the list created above
models.append(('LinearRegression',LinearRegression()))
models.append(('knn',KNeighborsRegressor(n_neighbors=n_neighbors)))
models.append(('SVR',SVR()))
models.append(("decision_tree",DecisionTreeRegressor()))

```

```python
# Evaluating Each model

#importing library
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score


names=[]
predictions=[]
error='neg_mean_squared_error'

for name,model in models:
    fold=KFold(n_splits=10,random_state=None)
    result=cross_val_score(model,x_train,y_train,cv=fold,scoring=error)
    predictions.append(result)
    names.append(name)
    result="%s : %f (%f)"%(name,result.mean(),result.std())
    print(result)
```

```python
 # Visualizing the Model accuracy
fig=plt.figure()
fig.suptitle("Comparing Algorithms")
axis=fig.add_subplot(111)
plt.boxplot(predictions)
axis.set_xticklabels(names)
plt.xticks(rotation='90')
plt.show()
```

```python
# Ensemble and Boosting algorithm to improve performance

#Ensemble
# Boosting methods
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
# Bagging methods
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor

ensembles=[]

#appending models to the list created above
ensembles.append(('AdaBoostRegressor',AdaBoostRegressor()))
ensembles.append(('GradientBoostingRegressor',GradientBoostingRegressor()))
ensembles.append(('RandomForestRegressor',RandomForestRegressor()))
ensembles.append(("ExtraTreesRegressor",ExtraTreesRegressor()))
```

```python
# Evaluating Each model


names=[]
results=[]

for name,model in ensembles:
    fold=KFold(n_splits=10,random_state=None)
    result=cross_val_score(model,x_train,y_train,cv=fold,scoring=error)
    results.append(result)
    names.append(name)
    result="%s : %f (%f)"%(name,result.mean(),result.std())
    print(result)
```

```python
# Visualizing the Model accuracy
fig=plt.figure()
fig.suptitle("Comparing ensembled Algorithms")
axis=fig.add_subplot(111)
plt.boxplot(results)
axis.set_xticklabels(names)
plt.xticks(rotation='90')
plt.show()
```
