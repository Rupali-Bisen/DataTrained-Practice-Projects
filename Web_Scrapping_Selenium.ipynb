### 1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
url1='https://www.naukri.com/'
driver.get(url1)

# Finding web element for search bar
search_bar=driver.find_element_by_id('qsb-keyword-sugg')
search_bar.send_keys('Data Analyst')

## Finding element for location bar
location_bar=driver.find_element_by_id('qsb-location-sugg')
location_bar.send_keys('Bangalore')


# clicking on search button using xpath function
search_btn1=driver.find_element_by_xpath("//button[@class='btn']")
search_btn1.click()

# Extracting job title
job_titles=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
Job_titles=[]
for i in job_titles:
    Job_titles.append(i.text)  

#Extracting job location
job_loc=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]')
Location=[]
for i in job_loc:
    Location.append(i.text)

#Extracting company name
comp_name=driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')
Company_name=[]
for i in comp_name:
    Company_name.append(i.text)
    
#Extracting Experience
exp=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi salary"]')
Experience=[]
for i in exp:
    Experience.append(i.text)

#create dataframe
Job_Data_Analyst=pd.DataFrame({})
Job_Data_Analyst['Designation']=Job_titles[:10]
Job_Data_Analyst['Location']=Location[:10]
Job_Data_Analyst['Company_name']=Company_name[:10]
Job_Data_Analyst['Experience']=Experience[:10]

Job_Data_Analyst
```

### 2. Write a python program to scrape data for “Data Scientist” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, full job-description.

```python
url2='https://www.naukri.com/'
driver.get(url2)
time.sleep(5)

# Finding web element for search bar
search_bar_DS=driver.find_element_by_id('qsb-keyword-sugg')
search_bar_DS.send_keys('Data Scientist')

## Finding element for location bar
location_bar_DS=driver.find_element_by_id('qsb-location-sugg')
location_bar_DS.send_keys('Bangalore')


# clicking on search button using xpath function
search_btn2=driver.find_element_by_xpath("//button[@class='btn']")
search_btn2.click()

# Extracting job title
job_titles_DS=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
Job_titles=[]
for i in job_titles_DS:
    Job_titles.append(i.text)  

#Extracting job location
job_loc_DS=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]')
Location=[]
for i in job_loc_DS:
    Location.append(i.text)

#Extracting company name
comp_name_DS=driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')
Company_name=[]
for i in comp_name_DS:
    Company_name.append(i.text)
    
#Extracting job description
urls=[]
for i in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):
    urls.append(i.get_attribute('href'))
    
Job_Desc=[]
for j in urls[0:10]:
    
    try:
        driver.get(j)
        description=driver.find_element_by_xpath("//section[@class='job-desc']").text
        Job_Desc.append(description)
    except NoSuchElementException:
        Job_Desc.append("Not Available")

#create dataframe
Job_Data_Scientist=pd.DataFrame({})
Job_Data_Scientist['Designation']=Job_titles[:10]
Job_Data_Scientist['Location']=Location[:10]
Job_Data_Scientist['Company_name']=Company_name[:10]
Job_Data_Scientist['Detailed_Description']=Job_Desc[:10]

Job_Data_Scientist
```

#### Q6 :Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes : Brand, Product Description, Price, Discount %

```python
driver.get('https://www.flipkart.com/')   
time.sleep(5)


#Navigating search bar
driver.find_element_by_xpath('//input[@class="_3704LK"]').send_keys('Sunglasses')
driver.find_element_by_xpath("//button[@class='L0Z3Pu']").click()
time.sleep(5)

#Extracting url
urls=[]
for i in driver.find_elements_by_xpath("//a[@class='_1LKTO3']"):
    urls.append(i.get_attribute('href'))

#Creating empty list
Brand=[]
Product_description=[]
Price=[]
Discount=[]

#Scrapping data
for j in urls[0:3]:
    driver.get(j)
    for i in driver.find_elements_by_xpath('//div[@class="_2WkVRV"]'):
        Brand.append(i.text.replace('\n',''))
Brand

for j in urls[0:3]:
    driver.get(j)
    for i in driver.find_elements_by_xpath('//a[@class="IRpwTa"]'):
        Product_description.append(i.text.replace('\n',''))
Product_description

for j in urls[0:3]:
    driver.get(j)
    for i in driver.find_elements_by_xpath('//div[@class="_30jeq3"]'):
        Price.append(i.text.replace('\n',''))
Price

for j in urls[0:3]:
    driver.get(j)
    try:
        for i in driver.find_elements_by_xpath('//div[@class="_3Ay6Sb"]'):
            Discount.append(i.text.replace('\n',''))
    except:
        Discount.append("NA")
Discount

#create dataframe
Flipkart_Sunglasses=pd.DataFrame({})
Flipkart_Sunglasses['Brand']=Brand[:100]
Flipkart_Sunglasses['Product_description']=Product_description[:100]
Flipkart_Sunglasses['Price']=Price[:100]
Flipkart_Sunglasses['Discount']=Discount[:100]

Flipkart_Sunglasses
```

### Q7: Scrape 100 reviews data from flipkart.com for iphone11 phone.

```python
driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-earpods-power-adapter/product-reviews/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace=FLIPKART&page=2')   
time.sleep(5)

#Creating empty list
phn_rating=[]
Review_summary=[] 
Full_review=[]


#Scrapping data
j=0
while j<10:
    try:
        for i in driver.find_elements_by_xpath('//div[@class="_3LWZlK _1BLPMq"]'):
            phn_rating.append(i.text)
    except:
        phn_rating.append('Not Available')
        
    try:
        for i in driver.find_elements_by_xpath('//p[@class="_2-N8zT"]'):
            Review_summary.append(i.text)
    except:
        Review_summary.append('Not Available')
        
    
    try:
        for i in driver.find_elements_by_xpath('//div[@class="t-ZTKy"]'):
            Full_review.append(i.text)
    except:
        Full_review.append('Not Available')
   
    #Extracting next page
    driver.find_element_by_xpath('//a[@class="_1LKTO3"]').click()
    time.sleep(10)
    j+=1
    time.sleep(10)
    
#create dataframe
iphone11_review=pd.DataFrame({})
iphone11_review['Rating']=phn_rating[:100]
iphone11_review['Review_summary']=Review_summary[:100]
iphone11_review['Full_review']=Full_review[:100]

iphone11_review   
```

#### Q9: Go to the link - https://www.myntra.com/shoesSet Price filter to “Rs. 6649 to Rs. 13099” , Color filter to “Black”, And then scrape First 100 shoes data
