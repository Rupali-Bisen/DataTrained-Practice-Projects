### 1. Scrape the details of most viewed videos on YouTube from Wikipedia: 

```python
url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
```

```python
rank=[]
name=[]
artist=[]
date=[]
views=[]
for k in range(1,31):
    try:
        for i in driver.find_elements_by_xpath("/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr[{}]/td[1]".format(k)):
            rank.append(i.text)
    except NoSuchAttributeException as e:
        rank.append('--')
        
    
    try:
        for i in driver.find_elements_by_xpath("/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr[{}]/td[2]".format(k)):
            name.append(i.text)
    except NoSuchAttributeException as e:
        name.append('--')
        
        
    try:
        for i in driver.find_elements_by_xpath("/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr[{}]/td[3]".format(k)):
            artist.append(i.text)
    except NoSuchAttributeException as e:
        artist.append('--')
    try:
        for i in driver.find_elements_by_xpath("/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr[{}]/td[4]".format(k)):
            views.append(i.text)
    except NoSuchAttributeException as e:
        views.append('--')
    try:
        for i in driver.find_elements_by_xpath("/html/body/div[3]/div[3]/div[5]/div[1]/table[3]/tbody/tr[{}]/td[5]".format(k)):
            date.append(i.text)
    except NoSuchAttributeException as e:
        date.append('--')
```

```python
df_wiki=pd.DataFrame({'Rank':rank,
                  'Name':name,
                  'Artist':artist,
                  'Upload date':date,
                  'Views':views})
df_wiki
```

### 2. Scrape the details team India’s international fixtures from bcci.tv. 

```python
driver.get('https://www.bcci.tv/')
```

```python
#scrapping  international fixture page
driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover']").click()
driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]/div/ul/li[1]/a').click()
```

```python
#creating empty list

match_title=[]
series=[]
place=[]
date=[]
time=[]


#scrapping match title
for i in driver.find_elements_by_xpath('//strong[@class="fixture__name fixture__name--with-margin"]'):
    match_title.append(i.text.replace('\n',' '))
match_title

#scrapping date
for i in driver.find_elements_by_xpath('//div[@class="fixture__full-date"]'):
    date.append(i.text.replace('\n',' '))
date

#scrapping time
for i in driver.find_elements_by_xpath('//span[@class="fixture__time"]'):
    time.append(i.text.replace('\n',' '))
time
#scrapping place
for i in driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]'):
    place.append(i.text.replace('\n',' '))
place
#scrapping series
for i in driver.find_elements_by_xpath('//span[@class="u-unskewed-text fixture__format"]'):
    series.append(i.text.replace('\n',' '))
series


#creating dataframe
bcci=pd.DataFrame({})
bcci['match_title']=match_title
bcci['date']=date
bcci['time']=time
bcci['place']=place
bcci['series']=series
bcci
```

### 10. Details of Datasets from UCI machine learning repositories

```python
driver.get('https://archive.ics.uci.edu/')
```

```python
driver.find_element_by_xpath("//span[2][@class='whitetext']/a").click()
```

```python
#creating empty list
data=[]
typ=[]
task=[]
attribute=[]
number=[]
n_attribute=[]
year=[]
for i in driver.find_elements_by_xpath("//p[@class='normal']/b/a"):
    data.append(i.text)
for k in range(2,590):
    for i in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[2]/p".format(k)):
        typ.append(i.text)
    for i in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[3]/p".format(k)):
        task.append(i.text)
    for i in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[4]/p".format(k)):
        attribute.append(i.text)
    for i in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[5]/p".format(k)):
        number.append(i.text)
    for i in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[6]/p".format(k)):
        n_attribute.append(i.text)
    for i in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{}]/td[7]/p".format(k)):
        year.append(i.text)
```

```python
df_repository=pd.DataFrame({'Dataset Name':data,
                            'Data Type':typ,
                            'Task':task,
                            'Attribute type':attribute,
                            'No of instances':number,
                            'Number of Attribute':n_attribute,
                            'Year':year})
df_repository
```
